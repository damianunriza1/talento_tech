```python
import nltk
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import os
import re

# Inicializar el lematizador y stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

# Función para tokenizar el texto
def tokenize(text):
    text = text.lower()
    words = re.findall(r'\b\w+\b', text)
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return words

# Función para cargar datos desde una carpeta
def load_data_from_folder(folder_path, label):
    data = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
                text = file.read()
                data.append((text, label))
    return data

# Cargar datos de las carpetas positivas y negativas
positive_folder = "txt_sentoken/pos"  # Reemplaza con la ruta a la carpeta de opiniones positivas
negative_folder = "txt_sentoken/neg"  # Reemplaza con la ruta a la carpeta de opiniones negativas

positive_data = load_data_from_folder(positive_folder, "positive")
negative_data = load_data_from_folder(negative_folder, "negative")

data = positive_data + negative_data

# Dividir el conjunto de datos en entrenamiento y prueba
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)

# Función para convertir el texto en diccionario de características
def extract_features(text):
    words = tokenize(text)
    return {word: True for word in words}

# Crear conjuntos de características para entrenamiento y prueba
train_set = [(extract_features(text), label) for text, label in train_set]
test_set = [(extract_features(text), label) for text, label in test_set]

# Entrenar el clasificador Naive Bayes utilizando NLTK
nltk_classifier = NaiveBayesClassifier.train(train_set)

# Evaluar la precisión del modelo Naive Bayes
nltk_accuracy = nltk.classify.accuracy(nltk_classifier, test_set)
print("Naive Bayes Accuracy:", nltk_accuracy)

# Hacer predicciones utilizando el conjunto de prueba Naive Bayes
nltk_test_texts, nltk_test_labels = zip(*test_set)
nltk_predictions = [nltk_classifier.classify(features) for features in nltk_test_texts]

# Calcular y mostrar la matriz de confusión para Naive Bayes
nltk_conf_matrix = confusion_matrix(nltk_test_labels, nltk_predictions, labels=['positive', 'negative'])
nltk_disp = ConfusionMatrixDisplay(nltk_conf_matrix, display_labels=['positive', 'negative'])
nltk_disp.plot(cmap=plt.cm.Blues)
plt.title("Naive Bayes Confusion Matrix")
plt.show()

# Vectorizar los textos para modelos scikit-learn
vectorizer = CountVectorizer(tokenizer=tokenize)
X_train = vectorizer.fit_transform([text for text, label in train_set])
X_test = vectorizer.transform([text for text, label in test_set])
y_train = [label for text, label in train_set]
y_test = [label for text, label in test_set]

# Entrenar el clasificador SVM
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X_train, y_train)

# Evaluar la precisión del modelo SVM
svm_predictions = svm_classifier.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)
print("SVM Accuracy:", svm_accuracy)

# Calcular y mostrar la matriz de confusión para SVM
svm_conf_matrix = confusion_matrix(y_test, svm_predictions, labels=['positive', 'negative'])
svm_disp = ConfusionMatrixDisplay(svm_conf_matrix, display_labels=['positive', 'negative'])
svm_disp.plot(cmap=plt.cm.Blues)
plt.title("SVM Confusion Matrix")
plt.show()

# Entrenar el clasificador Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Evaluar la precisión del modelo Random Forest
rf_predictions = rf_classifier.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)
print("Random Forest Accuracy:", rf_accuracy)

# Calcular y mostrar la matriz de confusión para Random Forest
rf_conf_matrix = confusion_matrix(y_test, rf_predictions, labels=['positive', 'negative'])
rf_disp = ConfusionMatrixDisplay(rf_conf_matrix, display_labels=['positive', 'negative'])
rf_disp.plot(cmap=plt.cm.Blues)
plt.title("Random Forest Confusion Matrix")
plt.show()
```
