```python
import pandas as pd
from sklearn.datasets import load_diabetes

# Cargar los datos de diabetes
diabetes_data = load_diabetes()

# Crear un DataFrame para las características (X)
X = pd.DataFrame(diabetes_data.data, columns=diabetes_data.feature_names)

# Crear un DataFrame para la variable objetivo (y)
y = pd.DataFrame(diabetes_data.target, columns=['target'])

# Combinar X e y en un solo DataFrame
df = pd.concat([X, y], axis=1)

# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

Select just one feature to target for this exercise

```python
# Selecting the 3rd feature
X = X[:, 2]
print(X.shape)
print(X[0])
X
```

```python
#Reshaping to get a 2D array
X = X.reshape(-1, 1)
print(X.shape)
print(X)
```

Split the training and test data for both `X` and `y`

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.20)
```

Select the model and fit it with the training data

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Use test data to predict a line

```python
y_pred = model.predict(X_test)
```

Display the results in a plot

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```

```python
# Calcular el Error Cuadrático Medio (MSE)
mse = mean_squared_error(y_test, y_pred)

print("MSE:", mse)
```

```python
# Calcular el Error absoluto medio (MAE)
mae = np.mean(np.abs(y_test - y_pred))

print("MAE:", mae)
```

```python
#Calcular el Error la raíz de MSE
rmse = np.sqrt(mse)

print("RMSE:", rmse)
```

```python
import matplotlib.pyplot as plt

# Calcular los errores
errors = y_test - y_pred

# Graficar los errores
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, errors, color='blue', alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.title('Gráfico de dispersión de errores')
plt.xlabel('Valores predichos')
plt.ylabel('Errores')
plt.grid(True)
plt.show()
```
