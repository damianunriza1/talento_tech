```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name = "microsoft/DialoGPT-large"  # Puedes usar 'medium' o 'large' para modelos más grandes
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def generar_respuesta(input_text, chat_history_ids=None):
    # Tokenizar la entrada
    new_input_ids = tokenizer.encode(input_text + tokenizer.eos_token, return_tensors='pt')

    # Agregar la entrada a la historia del chat
    bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids

    # Generar la respuesta
    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # Decodificar la respuesta y eliminar los tokens especiales
    respuesta = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

    return respuesta, chat_history_ids
```

```python
chat_history_ids = None
print("¡Hola! Soy un chatbot con IA. Escribe algo para empezar a conversar (escribe 'adiós' para terminar la conversación).")

while True:
    user_input = input("Tú: ")
    if user_input.lower() == "adiós":
        print("Chatbot: Hasta luego, ¡cuídate!")
        break

    # Generar respuesta
    respuesta, chat_history_ids = generar_respuesta(user_input, chat_history_ids)
    print("Chatbot:", respuesta)
```
