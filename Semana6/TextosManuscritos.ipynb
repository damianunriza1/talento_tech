```python
import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Cargar el conjunto de datos MNIST
mnist = keras.datasets.mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()

# Normalizar los valores de píxeles del tren y prueba las imágenes.
training_images = training_images / 255.0
test_images = test_images / 255.0

# Separar datos de entrenamiento en entrenamiento y validación
train_images, val_images, train_labels, val_labels = train_test_split(training_images, training_labels, test_size=0.2, random_state=42)

# Función para construir, compilar y entrenar el modelo con hiperparámetros específicos
def train_and_evaluate_model(layer1_size, layer2_size, epochs, l2_reg):
    # Construir el modelo de clasificación
    model = keras.models.Sequential([
        keras.layers.Flatten(input_shape=(28, 28)),
        keras.layers.Dense(layer1_size, activation='tanh', kernel_regularizer=keras.regularizers.l2(l2_reg)),
        keras.layers.Dense(layer2_size, activation='tanh', kernel_regularizer=keras.regularizers.l2(l2_reg)),
        keras.layers.Dense(10, activation='softmax')
    ])

    # Compilar el modelo
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Entrenar el modelo con datos de entrenamiento y validación
    history = model.fit(train_images, train_labels, epochs=epochs, validation_data=(val_images, val_labels), verbose=0)

    # Crear DataFrame a partir del historial
    history_df = pd.DataFrame(history.history)

    # Evaluar el modelo en el conjunto de prueba
    test_loss, test_accuracy = model.evaluate(test_images, test_labels, verbose=0)

    # Predecir etiquetas para el conjunto de prueba
    test_predictions = model.predict(test_images)
    test_predictions_labels = np.argmax(test_predictions, axis=1)

    # Calcular la matriz de confusión
    conf_matrix = confusion_matrix(test_labels, test_predictions_labels)

    return history_df, conf_matrix

# Diferentes configuraciones de hiperparámetros para probar
hyperparameters = [
    {'layer1_size': 128, 'layer2_size': 64, 'epochs': 10, 'l2_reg': 0.01},
    {'layer1_size': 256, 'layer2_size': 128, 'epochs': 10, 'l2_reg': 0.01},
    {'layer1_size': 128, 'layer2_size': 64, 'epochs': 20, 'l2_reg': 0.001},
    {'layer1_size': 256, 'layer2_size': 128, 'epochs': 20, 'l2_reg': 0.001}
]

# Graficar los resultados
plt.figure(figsize=(12, 8))

for params in hyperparameters:
    history_df, conf_matrix = train_and_evaluate_model(params['layer1_size'], params['layer2_size'], params['epochs'], params['l2_reg'])
    plt.plot(history_df['loss'], label=f"Train Loss: L1={params['layer1_size']}, L2={params['layer2_size']}, Epochs={params['epochs']}, L2={params['l2_reg']}")
    plt.plot(history_df['val_loss'], label=f"Val Loss: L1={params['layer1_size']}, L2={params['layer2_size']}, Epochs={params['epochs']}, L2={params['l2_reg']}", linestyle='dashed')

plt.title('Training and Validation Loss for Different Hyperparameters')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar las matrices de confusión
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

for i, params in enumerate(hyperparameters):
    _, conf_matrix = train_and_evaluate_model(params['layer1_size'], params['layer2_size'], params['epochs'], params['l2_reg'])
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=np.arange(10))
    disp.plot(ax=axes[i])
    axes[i].set_title(f"L1={params['layer1_size']}, L2={params['layer2_size']}, Epochs={params['epochs']}, L2={params['l2_reg']}")

plt.tight_layout()
plt.show()

```
