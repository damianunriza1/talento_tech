##Regresión Lineal

```python
import numpy as np
import pandas as pd
```

```python

import matplotlib.pyplot as plt
```

```python
x = [1, 2, 2, 3, 4, 4, 5, 6, 8 , 9 ,12 ,14 ,20]
y = [2, 3, 4, 4, 4, 6 , 5 ,7, 6 ,7, 8, 9, 12]

df = pd.DataFrame({'x': x, 'y': y})

print(df)
```

```python
df['xy']=df['x']*df['y']
print(df)
```

```python
df['xx']=df['x']*df['x']
print(df)
```

```python

mean_x = np.mean(x)
mean_y = np.mean(y)

numerador = np.sum((x - mean_x) * (y - mean_y))
denominador = np.sum((x - mean_x) ** 2)
m = numerador / denominador

print(m)
```

```python
b= mean_y - m*mean_x
print(b)
```

```python
# Crear la recta de regresión lineal
x_values = np.array(x)
y_values = m * x_values + b

# Graficar los datos y la recta de regresión lineal
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x_values, y_values, color='red', label='Recta de regresión lineal')
plt.xlabel('Variable independiente (x)')
plt.ylabel('Variable dependiente (y)')
plt.title('Regresión lineal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['diferencia']=y_values-df['y']
print(df)
```

```python
#Error cuadratico medio
df['diferencia_pred2']=df['diferencia']**2
print(df)
```

```python
rmse=np.sqrt(df['diferencia_pred2'].mean())
print(rmse)
```
