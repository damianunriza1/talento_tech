```python
import cv2
from matplotlib import pyplot as plt

# Ruta del archivo
image =cv2.imread( 'gatos3.png')


```

```python
height, width =image.shape[:2]
center=(width//2,height//2)
```

```python

cv2_imshow(image)
angle=45
M=cv2.getRotationMatrix2D(center,angle,1)
rotated_image=cv2.warpAffine(image,M,(width,height))
cv2_imshow(rotated_image)
```

```python
smoothed=cv2.GaussianBlur(image,(5,5),0)
cv2_imshow(smoothed)
```

```python
kernel=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
sharpened=cv2.filter2D(image,-1,kernel)
cv2_imshow(sharpened)

```

```python
sobelx=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)
spbely=cv2.Sobel(image,cv2.CV_64F,0,1,ksize=5)

edges=cv2.magnitude(sobelx,spbely)
edges=cv2.normalize(edges,None,0,255,cv2.NORM_MINMAX)
edges=edges.astype(np.uint8)
cv2_imshow(edges)
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
faceClassif=cv2.CascadeClassifier( 'haarcascade_frontalcatface.xml')
```

```python
image = cv2.imread('gatos3.png')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray)
```

```python
faces = faceClassif.detectMultiScale(gray,scaleFactor=1.1, minNeighbors=5, minSize=(30,30), maxSize=(1000,1000))
print(faces)
```

```python
for(x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows
```
